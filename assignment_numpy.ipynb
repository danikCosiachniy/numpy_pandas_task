**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
arr = np.random.randint(0,15,10)
print(arr)
mask = (arr > 3) & (arr < 8)
arr[mask] *= -1
print(arr)
```

**2.** Заменить максимальный элемент случайного массива на 0

```python
arr = np.random.randint(0,15,10)
print(arr)
arr[arr == arr.max()] = 0
print(arr)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
arr = np.random.randint(10, 50, size=(2, 3))
arrays = [row for row in arr]
combinations = np.stack(np.meshgrid(*arrays), axis=-1).reshape(-1, arr.shape[0])
print(combinations)
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.random.randint(10, 50, size=(8, 3))
print(A)
B = np.random.randint(10, 50, size=(2, 2))
print(B)
mask1 = np.isin(A, B[0]).any(axis=1)
mask2 = np.isin(A, B[1]).any(axis=1)
final_mask = mask1 & mask2
result = A[final_mask]
print(result)
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
matrix = np.random.randint(0, 4, size=(10, 3))
print(matrix)
ranges = np.ptp(matrix, axis=1)
mask = ranges > 0
result = matrix[mask]
print(result)
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
arr = np.random.randint(0,4, size=(4,3))
print(arr)
arr_unique = np.unique(arr, axis=0)
print(arr_unique)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
X_list = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]

def prod_diag_non_numpy(matrix):
    prod = 1
    d_len = min(len(matrix), len(matrix[0]))

    for i in range(d_len):
        if matrix[i][i] != 0:
            prod *= matrix[i][i]
    return prod
print(prod_diag_non_numpy(X_list))

X_numpy = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
result = np.diag(X_numpy)[np.diag(X_numpy) != 0].prod()
print(result)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
X_list = [2, 2, 2, 3, 3, 3, 5]
def count_in_list_not_numpy(arr):
    num = []
    count = []
    for i in range(len(arr)):
        if arr[i] not in num:
            num.append(arr[i])
            count.append(arr.count(arr[i]))
    return (num, count)
print(count_in_list_not_numpy(X_list))

x_numpy = np.array([2, 2, 2, 3, 3, 3, 5])

def rle_numpy(x):
  change_mask = np.append(True, x[1:] != x[:-1])
  run_starts_indices = np.where(change_mask)[0]
  values = x[run_starts_indices]
  counts = np.diff(np.append(run_starts_indices, x.size))
  
  return (values, counts)

print(rle_numpy(x_numpy))
        
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
import math

X_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9], [1, 1, 1]]
Y_list = [[0, 0, 0], [1, 2, 4]]

def cdist_non_numpy(X, Y):
  dist_matrix = []
  for x_vec in X:
    row = []
    for y_vec in Y:
      sum_of_squares = sum([(x_k - y_k)**2 for x_k, y_k in zip(x_vec, y_vec)])
      row.append(math.sqrt(sum_of_squares))
    dist_matrix.append(row)
  return dist_matrix

for row in cdist_non_numpy(X_list, Y_list):
    print([round(val, 2) for val in row])
    
X_np = np.array(X_list)
Y_np = np.array(Y_list)

def cdist_numpy(X, Y):
  sum_X_sq = np.sum(X**2, axis=1, keepdims=True)
  sum_Y_sq = np.sum(Y**2, axis=1)                
  neg_2XY = -2 * X @ Y.T                       
  dist_sq = sum_X_sq + neg_2XY + sum_Y_sq
  return np.sqrt(np.maximum(0, dist_sq))

print(np.round(cdist_numpy(X_np, Y_np), 2))
```
